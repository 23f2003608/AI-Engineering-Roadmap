###`Q-2:` There is an array of marks of 5 students in 4 subjects. Further you are asked to perform below task.
    i. Add marks every student of an extra subject in the same array.
    ii. Add two new students marks in respective 5 subjects.(one subject added in above task)
    iii. Add extra column with sum of all subjects(5-subjects) marks
    iv. Sort the array(non-ascending order) on total marks column--one added in above task. Show top 2 rows.

Note: Change dimension of arrays during concatenation or appending if required.

Given Array-
```
marks = [[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]]

extra_subject = [41, 87, 72, 36, 92]
#Two extra students record-
rec1 = [77, 83, 98, 95, 89]
rec2 = [92, 71, 52, 61, 53]
```

````python
# code here
marks = [[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]]

extra_subject = [41, 87, 72, 36, 92]
#Two extra students record-
rec1 = [77, 83, 98, 95, 89]
rec2 = [92, 71, 52, 61, 53]

# Step i
marks_with_extra = np.column_stack((marks, extra_subject))

# Step ii
all_marks = np.concatenate((marks_with_extra, [rec1, rec2]), axis=0)

# Step iii
marks_with_total = np.column_stack((all_marks, all_marks.sum(axis=1)))

# Step iv
sorted_marks = marks_with_total[marks_with_total[:, -1].argsort()[::-1]]
top2 = sorted_marks[:2]

print("After adding extra subject:\n", marks_with_extra)
print("\nAfter adding two extra students:\n", all_marks)
print("\nAfter adding total marks column:\n", marks_with_total)
print("\nAfter sorting on total marks:\n", sorted_marks)
print("\nTop 2 students:\n", top2)
````

### `Q-6:` How to convert an array of arrays into a flat 1d array?


````python
# These arrays are given.
arr1 = np.arange(3)
arr2 = np.arange(3,7)
arr3 = np.arange(7,10)
def convert(*arrays):
    merge = np.concatenate(arrays)
    return np.ravel(merge)

print(convert(arr1, arr2, arr3))
````

### `Q-9:` You are given a array ($arr ∈ [0, 1]$). First you have round off the elements upto 3 decimal places and compare that
- 0th percentile == minimum value of the array
- 100th percentile == maximum value of the array
- also find the difference betwen 51th percenile and 50th percentile values

````python
# code here
# Random array example
arr = np.random.rand(10)

# round to 3 decimals
arr_rounded = np.round(arr, 3)

# 0th and 100th percentile
p0 = np.percentile(arr_rounded, 0)
p100 = np.percentile(arr_rounded, 100)

print("Rounded array:", arr_rounded)
print("0th percentile == minimum value of the array??"," ", p0 == arr_rounded.min())
print("100th percentile == maximum value of the array??"," ", p100 == arr_rounded.max())

# Step 3: 50th & 51st percentile
p50 = np.percentile(arr_rounded, 50)
p51 = np.percentile(arr_rounded, 51)
diff = p51 - p50

print("50th percentile:", p50)
print("51st percentile:", p51)
print("Difference (51-50):", diff)
````
